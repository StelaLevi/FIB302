```python
v_0 = 20
theta = np.radians(45)
g = 9.8
tempo = 2


posicao_horizontal = v_0 * np.cos(theta) * tempo
posicao_vertical = v_0 * np.sin(theta) * tempo - 0.5 * g * tempo**2

print("Posição horizontal:", posicao_horizontal, "metros")
print("Posição vertical:", posicao_vertical, "metros")
```

```python
import numpy as np
import matplotlib.pyplot as plt

tempo = np.linspace(0, 2*np.pi, 1000)
A = 1
omega = 2*np.pi

deslocamento = A * np.cos(omega * tempo)

plt.plot(tempo, deslocamento)
plt.title('Deslocamento em Movimento Harmônico Simples')
plt.xlabel('Tempo (s)')
plt.ylabel('Deslocamento (m)')
plt.grid(True)
plt.show()
```
